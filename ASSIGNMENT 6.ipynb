```python
secret_word = "python"
guess = ""

while guess != secret_word:
    guess = input("Guess the word: ")

print("🎉 You guessed it!")
```

# QUESTION 2

```python
users = ["Ali", "Sara", "John", "Zara", "Mike"]

for user in users:
    print("Email sent to", user)

    if user == "John":
        print("Limit reached. Stop sending emails.")
        break
```

# QUESTION 3

```python
temps = [22, 25, 30, 45, 28]
critical = 40

for temp in temps:
    if temp >= critical:
        print("Critical temperature reached! Stop monitoring.")
        break

    print("Temperature:", temp)
```

# QUESTION 4

```python
cart = [100, 250, 50, 300]

total = 0

for price in cart:
    total = total + price

print("Total Price:", total)
```

# QUESTION 5

```python
customers = [
    {"name": "Ali", "paid": True},
    {"name": "Sara", "paid": False},
    {"name": "John", "paid": True}
]

for customer in customers:
    if customer["paid"] == False:
        continue

    print("Access granted to:", customer["name"])
```

# QUESTION 7

```python
emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

domains = []

for email in emails:
    domain = email.split("@")[1]
    domains.append(domain)

print("domains:", domains)
```

# QUESTION 8

```python
tasks = [
    {"task": "Pay bills", "status": "done"},
    {"task": "Submit report", "status": "pending"},
    {"task": "Clean house", "status": "pending"}
]

for task in tasks:
    if task["status"] == "pending":
        print("Overdue:", task["task"])
```

# QUESTION 9

```python
while True:
    number = int(input("Enter a number: "))

    if number < 0:
        print("Negative number entered. Loop stopped!")
        break

    print("You entered:", number)
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    if day == "Saturday" or day == "Sunday":
        continue

    print("Workday:", day)
```
