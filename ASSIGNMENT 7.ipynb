```python
def currency_converter(amount, exchange_rate=280):
    converted_amount = amount * exchange_rate
    return f"{amount} USD is equivalent to {converted_amount} PKR"

# Function call
result = currency_converter(100)

print(result)
```

# QUESTION 5

```python
def calculate_rectangle(length, width):

    # Nested function to calculate area
    def calculate_area():
        return length * width

    # Nested function to calculate perimeter
    def calculate_perimeter():
        return 2 * (length + width)

    # Return both values
    return calculate_area(), calculate_perimeter()


# Function call
area, perimeter = calculate_rectangle(6, 4)

print("Area:", area)
print("Perimeter:", perimeter)
```

# QUESTION 6

```python
# Original list
numbers = [3, 5, 7, 9, 10, 12, 15, 18, 20, 22, 25, 30]

# List comprehension
result = [num for num in numbers if num % 3 == 0 or num % 5 == 0]

print("Original List:", numbers)
print("Numbers divisible by 3 or 5:", result)
```

# QUESTION 7

```python
balance = 1000

while True:
    try:
        amount = float(input("Enter amount to withdraw: "))

        if amount <= 0:
            print("Amount must be more than zero.")
            continue

        elif amount > balance:
            print("Insufficient balance.")
            continue

        else:
            balance = balance - amount
            print("Withdrawal successful.")
            print("Remaining balance:", balance)
            break

    except ValueError:
        print("Please enter a valid number.")
        continue

print("Thank you for using the ATM.")
```

# QUESTION 8

```python
available_seats = 5

while True:
    try:
        tickets = int(input("How many tickets do you want to book? "))

        if tickets <= 0:
            print("You must book at least 1 ticket.")
            continue

        elif tickets > available_seats:
            print("Not enough seats available.")
            continue

        else:
            print(tickets, "ticket(s) booked successfully.")
            break

    except ValueError:
        print("Please enter a valid number.")
        continue

print("Thank you for booking with us.")
```

```python
# Define the class
class Light:

    # Method to turn on the light
    def turn_on(self):
        print("The light is now ON.")

    # Method to check the light status
    def check_status(self):
        print("The light is bright and working.")


# Create an object
bulb = Light()

# Call the methods
bulb.turn_on()
bulb.check_status()
```

# QUESTION 10

```python
# Define the class
class Student:

    # Constructor
    def __init__(self, name, grade, age):
        self.name = name
        self.grade = grade
        self.age = age

    # Method to introduce the student
    def introduce(self):
        print(f"My name is {self.name}, I am {self.age} years old.")

    # Method to show the grade
    def show_grade(self):
        print(f"My grade is {self.grade}.")


# Create an object
student1 = Student("Alice", "A", 16)

# Call the methods
student1.introduce()
student1.show_grade()
```
