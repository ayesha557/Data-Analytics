# QUESTION 1

```python
# Nested dictionary
products = {
    101: {"name": "Laptop", "category": "Electronics", "price": 1200, "stock": 50},
    102: {"name": "Shirt", "category": "Apparel", "price": 25, "stock": 200},
    103: {"name": "Coffee Maker", "category": "Home Appliances", "price": 80, "stock": 30}
}

# Increase Shirt's stock from 200 to 250
products[102]["stock"] = 250

# Display updated dictionary
print(products)

# Add a new product
products[104] = {
    "name": "Smartphone",
    "category": "Electronics",
    "price": 800,
    "stock": 100
}

# Display dictionary after adding new product
print(products)
```

# QUESTION 4

```python
# Create sets
course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

# Difference
difference = course_a.difference(course_b)

# Symmetric Difference
sym_diff = course_a.symmetric_difference(course_b)

# Display results
print("Students in Course A but not in Course B:", difference)
print("Students only in one of the two courses:", sym_diff)
```

# QUESTION 6

```python
# Dictionary
student_details = {
    "Ali": {"age": 20, "grade": "A", "city": "Karachi"},
    "Sara": {"age": 21, "grade": "B", "city": "Lahore"},
    "Omar": {"age": 19, "grade": "A", "city": "Islamabad"},
    "Hina": {"age": 22, "grade": "C", "city": "Faisalabad"}
}

# Add new student
student_details.update({
    "Zara": {"age": 20, "grade": "B", "city": "Multan"}
})

# Remove Omar
student_details.pop("Omar")

# Display dictionary
print(student_details)
```

# QUESTION 10

```python
# Sets
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

# Common elements
print("Common elements:", A.intersection(B))

# Elements only in Set A
print("Only in Set A:", A.difference(B))

# Elements only in Set B
print("Only in Set B:", B.difference(A))
```
