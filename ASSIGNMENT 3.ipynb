# QUESTION 5

```python
text = "hello python world"

# Replace spaces with underscores
modified = text.replace(" ", "_")

# Split the modified string
words = modified.split("_")

# Join with hyphens
joined = "-".join(words)

print("Modified:", modified)
print("List:", words)
print("Joined:", joined)
```

# QUESTION 7

```python
code = "SKU-12345-XYZ"

# First position of dash
first_dash = code.find("-")

# Count dashes
dash_count = code.count("-")

# Extract numeric part
number = code.replace("SKU-", "").replace("-XYZ", "")

print("First dash position:", first_dash)
print("Number of dashes:", dash_count)
print("Numeric part:", number)
```

# QUESTION 9

```python
text = " Data Analysis with Python is very powerful "

# Remove extra spaces
text = text.strip()

# Convert to title case
title_text = text.title()

# Replace Powerful with Important
updated_text = title_text.replace("Powerful", "Important")

# Find position of Python
position = updated_text.find("Python")

print("Stripped:", text)
print("Title Case:", title_text)
print("Updated:", updated_text)
print("Position of Python:", position)
```

# QUESTION 10

```python
data = [12, 45, 2, 67, 34, 89, 23, 90]

# Sort ascending
data.sort()

# Remove highest value
data.remove(max(data))

# Insert 50 at index 3
data.insert(3, 50)

# Reverse list
data.reverse()

print(data)
```
